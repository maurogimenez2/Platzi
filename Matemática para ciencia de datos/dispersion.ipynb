```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.sample(5)
```

```python
sns.scatterplot(x = iris["sepal_length"], y= iris["petal_length"], hue = iris["species"]) 
```

```python
sns.jointplot(x = iris["sepal_length"], y= iris["petal_length"], hue = iris["species"])
```

```python
sns.boxplot(x = iris["species"], y= iris["petal_length"])
```

```python
sns.barplot(x = iris["species"], y= iris["petal_length"])
```
